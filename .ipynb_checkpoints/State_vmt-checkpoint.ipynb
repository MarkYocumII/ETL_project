```python
import pandas as pd
import requests
from requests.exceptions import HTTPError
```

```python
#create excel file names for website download by looping through years and months

years = ["16", "17", "18"]
months = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
excel_files=[]

for months in months:
    for year in years:
        excel_files.append(year + months)
        
```

```python
#create empty list of monthly VMT (vehicle miles travelled) dataframes

month_dfs = []

#download each individual excel file via loop:
for file in excel_files:
    
    try:
        #try to download excel file, PAGG 6
        data_url = f"https://www.fhwa.dot.gov/policyinformation/travel_monitoring/{file}tvt/{file}tvt.xls"
        data = pd.read_excel(data_url, "Page 6", skiprows=7)
        
        #rename first column to "State"
        data = data.rename(columns={"Unnamed: 0": "State"})
        
        #extract only state, stations and CMILES (total vehicle miles travelled)
        data = data[["State", "STATIONS", "CMILES"]]
        data = data.dropna()
        
        #remove TOTALS and rename Stations and Miles columns
        data = data[data["State"] != "TOTALS"]
        data = data.rename(columns={"STATIONS": f"{file} Stations"})
        data = data.rename(columns={"CMILES": f"{file} Miles"})
        
        #set index to state and add monthly dataframe to list
        data = data.set_index("State")
        month_dfs.append(data)
    
    except:
        print(f"{file} not found ....skipping")
        continue

#merge entire list of individual monthly dataframes
state_vmt = pd.DataFrame()
state_vmt = pd.concat(month_dfs, axis=1, sort=True)
state_vmt = state_vmt.dropna()
state_vmt.head()
```
